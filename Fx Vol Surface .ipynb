```python
import numpy as np
import pandas as pd
import math as m
import pathlib 

input_dir = pathlib.Path('Input Files')

ATM = pd.read_excel(input_dir / 'FX vol surface.xls', sheet_name = 'ATM', na_filter = False,
                   parse_dates = ['Date'])
RR10 = pd.read_excel(input_dir / 'FX vol surface.xls', sheet_name = 'RR10', na_filter = False,
                    parse_dates = ['Date'])
FLY10 = pd.read_excel(input_dir / 'FX vol surface.xls', sheet_name = 'FLY10', na_filter = False,
                     parse_dates = ['Date'])
RR25 = pd.read_excel(input_dir / 'FX vol surface.xls', sheet_name = 'RR25', na_filter = False,
                    parse_dates = ['Date'])
FLY25 = pd.read_excel(input_dir / 'FX vol surface.xls', sheet_name = 'FLY25', na_filter = False,
                     parse_dates = ['Date'])
```

```python
Put10 = pd.DataFrame().reindex_like(ATM).dropna()
Put10['Date'] = ATM['Date']

Put10
```

```python
Put25 = pd.DataFrame().reindex_like(ATM).dropna()
Put25['Date'] = ATM['Date']
Call25 = pd.DataFrame().reindex_like(ATM).dropna()
Call25['Date'] = ATM['Date']
Call10 = pd.DataFrame().reindex_like(ATM).dropna()
Call10['Date'] = ATM['Date']
```

```python
for i in Put10.iloc[0:,1:]:
    Put10[i] = ATM[i] - 0.5 * RR10[i] + FLY10[i]
for i in Put25.iloc[0:,1:]:
    Put25[i] = ATM[i] - 0.5 * RR25[i] + FLY25[i]

for i in Call25.iloc[0:,1:]:
    Call25[i] = ATM[i] + 0.5 * RR25[i] + FLY25[i]
    
for i in Call10.iloc[0:,1:]:
    Call10[i] = ATM[i] + 0.5 * RR10[i] + FLY10[i]
```

```python
Put10.columns = Put10.columns.map(lambda x: 'Put10_' + str(x))

Put10
```

```python
ATM = ATM.set_index('Date')
Put25 = Put25.set_index('Date')
Call25 = Call25.set_index('Date')
Call10 = Call10.set_index('Date')
```

```python
ATM.columns = ATM.columns.map(lambda x: 'ATM_' + str(x))
Put25.columns = Put25.columns.map(lambda x: 'Put25_' + str(x))
Call25.columns = Call25.columns.map(lambda x: 'Call25_' + str(x))
Call10.columns = Call10.columns.map(lambda x: 'Call10_' + str(x))
```

```python
from functools import reduce

surface = [Put10, Put25,ATM, Call25, Call10]

fx_vol_surface = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), surface)
```
